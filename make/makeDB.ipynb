```python
import pandas as pd

df = pd.read_csv("winequality-red.csv")
df.head()
```

```python

df.reset_index(inplace=True)
df.head()
```

```python
df['index'] = df['index']+1
```

```python
def col_name(df):
    col_list =[]
    for name in df.columns:
        col_list.append(name.replace(' ',''))
    df.columns = col_list
    return df

def round_col(df):
    for i in df:
        # print(i)
        if(type(i) is not int):
            return round(i,2)

```

```python
for i in df:
    df.apply(round_col)
```

```python
df = col_name(df)
```

```python
col = list(df.columns)
col.remove('index')
```
